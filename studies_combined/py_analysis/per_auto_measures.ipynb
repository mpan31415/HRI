```python
from pandas import read_csv
from os import getcwd
import matplotlib.pyplot as plt
from scipy.stats import gaussian_kde
from numpy import vstack
from math import floor

STUDY1_RESULTS_PATH = getcwd() + "\\..\\..\\study1_results\\"
STUDY2_RESULTS_PATH = getcwd() + "\\..\\..\\study2_results\\"

measures = ["tlx_ave", "tapping_err_short", "tapping_err_long", "pupil_index", "mdmt_ave", "single_trust", "overall_traj_err"]

df1 = read_csv(STUDY1_RESULTS_PATH + "all_data.csv", index_col=False)
df2 = read_csv(STUDY2_RESULTS_PATH + "all_data.csv", index_col=False)

per_auto_list2 = df2['per_auto'].tolist()
```

## Against measures from study 1

```python
f = plt.figure()
f.set_figwidth(18)
f.set_figheight(6)

for i in range(len(measures)):

    measure = measures[i]
    measure_list = df1[measure].tolist()

    # scatter plot
    xy = vstack([per_auto_list2, measure_list])
    z = gaussian_kde(xy)(xy)

    plt.subplot(2, 4, i+1)
    plt.scatter(per_auto_list2, measure_list, c=z, s=20)
    # plt.xlabel("Perceived Autonomy")
    # plt.ylabel(measure)
    plt.title(measure)

plt.show()
```
