```python
from pandas import read_csv
from os import getcwd
import matplotlib.pyplot as plt
from scipy.stats import gaussian_kde
from numpy import vstack

STUDY1_RESULTS_PATH = getcwd() + "\\..\\..\\study1_results\\"
STUDY2_RESULTS_PATH = getcwd() + "\\..\\..\\study2_results\\"
```

#### Study 1 - Actual vs. Perceived Autonomy

```python
df1 = read_csv(STUDY1_RESULTS_PATH + "all_data.csv", index_col=False)

actual_auto_list = df1['autonomy'].tolist()
perceived_auto_list = df1['per_auto'].apply(lambda x: float(x/10)).tolist()

diff_list = [abs(perceived_auto_list[i] - actual_auto_list[i]) for i in range(len(actual_auto_list))]
diff_sum = sum(diff_list) / len(diff_list)

print("Average of differences = %.3f" % diff_sum)


# scatter plot
xy = vstack([actual_auto_list, perceived_auto_list])
z = gaussian_kde(xy)(xy)

plt.scatter(actual_auto_list, perceived_auto_list, c=z, s=100)
plt.show()
```
