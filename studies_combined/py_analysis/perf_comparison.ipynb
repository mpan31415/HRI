```python
from pandas import read_csv
from os import getcwd
from scipy.stats import f_oneway
import matplotlib.pyplot as plt

STUDY1_RESULTS_PATH = getcwd() + "\\..\\..\\study1_results\\"
STUDY2_RESULTS_PATH = getcwd() + "\\..\\..\\study2_results\\"

df1 = read_csv(STUDY1_RESULTS_PATH + "traj_csv_logs\\tap_impact.csv", index_col=False)
df2 = read_csv(STUDY2_RESULTS_PATH + "primary_task\\tap_impact.csv", index_col=False)
```

```python
exp1_baseline_errs = df1['base_err']
exp2_baseline_errs = df2['base_err']

res = f_oneway(exp1_baseline_errs, exp2_baseline_errs)
print("ANOVA results: %s", res)

plt.boxplot([exp1_baseline_errs, exp2_baseline_errs])
plt.show()
```
