```python
from pandas import read_csv
from scipy.stats import f_oneway
```

```python
df = read_csv("tap_impact.csv")
df
```

```python
baseline_err = df['base_err'].tolist()
round1_err = df['round1_err'].tolist()
zero_auto_err = df['no_auto_err'].tolist()

# did tapping negatively impact trajectory tracking? (compare baseline <-> zero_auto)
f_oneway(baseline_err, zero_auto_err)

######   TAPPING DIDN'T IMPACT THE PRIMARY TRAJECTORY TRACKING TASK!!!   ######
```

```python
baseline_ave = sum(baseline_err) / len(baseline_err)
print("Average error in baseline = %.3f" % baseline_ave)
```
